```python
def increase_age(age):
    return age + 5
obesity_data['Age'] = obesity_data['Age'].apply(increase_age)
```

```python
obesity_data.head()
```

```python
def decrease_weight(weight):
    return weight - 6

obesity_data['Weight'] = obesity_data['Weight'].apply(decrease_weight)
```

```python
obesity_data.head()
```

```python
obesity_data['obesity_level'] = obesity_data['Obesity'].apply(lambda x: 'Healthy' if x == 'Normal' else 'Unhealthy')
# Creating a new column based on a condition using a lambda function
```

```python
obesity_data.head()
```

```python
import numpy as np # type: ignore
obesity_data['New Age Category'] = np.where(obesity_data['Age'] < 30, 'Young',
                                           np.where(obesity_data['Age'] < 60, 'Middle-aged', 'Senior'))
```

```python
obesity_data.head()
```

```python
obesity_data.insert(2,'BMI',obesity_data['Weight'] / (obesity_data['Height']/100)**2,allow_duplicates=True)
```

```python
obesity_data.drop(columns='obesity_level')
```

```python
df1=pd.DataFrame({'id':[1,2,3,4,5],
    'Name':['Alice','Bob','Charlie','David','Eva'],
                  'Age':[25,30,35,40,45]})
df2=pd.DataFrame({'id':[3,4,5,6,7],'Name':['Frank','Grace','Hannah','Ian','Jack'],
                  'Age':[28,33,38,43,48]})

inner_merge = pd.merge(df1, df2, how='inner', on='id')#intersection of both df based on id col
inner_merge
```

```python
df1=pd.DataFrame({'id':[1,2,3,4,5],
    'Name':['Alice','Bob','Charlie','David','Eva'],
                  'Age':[25,30,35,40,45]})
df2=pd.DataFrame({'id':[ 3,4,5,6,7],'Name':['Frank','Grace','Hannah','Ian','Jack'],
                  'Age':[28,33,38,43,48]})#outer merge to get union from both df based on id col

outer_merge = pd.merge(df1, df2, how='outer', on='id')
outer_merge
```

```python
left_merge = pd.merge(df1, df2, how='left', on='id')
left_merge#left join to get all records from left df1 based on id col
```

```python
right_merge = pd.merge(df1, df2, how='right', on='id')
right_merge#right join to get all records from right df2 based on id col
```

```python
df3=pd.DataFrame({'id1':[1,2,3,4,5],
    'Name':['Alice','Bob','Charlie','David','Eva'],
                  'Age':[25,30,35,40,45]})
df4=pd.DataFrame({'id2':[ 3,4,5,6,7],'Name':['Frank','Grace','Hannah','Ian','Jack'],
                  'Age':[28,33,38,43,48]})#outer merge to get union from both df based on id col

outer_merge = pd.merge(df3, df4, how='inner', left_on='id1', right_on='id2')
outer_merge
```

```python
# create new dfs with a 'Course' column containing some missing/empty values
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Course': ['Math', 'Science', None, 'History', '']  # None and empty string as missing values
})
df2 = pd.DataFrame({
    'id': [3, 4, 5, 6, 7],
    'Name': ['Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Age': [28, 33, 38, 43, 48],
    'Course': ['Physics', '', 'English', None, 'Art']  # some missing/empty entries
})

# vertical concatenation and normalize empty strings to NaN so pandas considers them as missing
result = pd.concat([df1, df2], ignore_index=True)
 

result
```

```python
result.isna()
```

```python
result.notna()
```

```python
# create new dfs with a 'Course' column containing some missing/empty values
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Course': ['Math', 'Science', None, 'History', '']  # None and empty string as missing values
})
df2 = pd.DataFrame({
    'id': [3, 4, 5, 6, 7],
    'Name': ['Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Age': [28, 33, 38, 43, 48],
    'Course': ['Physics', '', 'English', None, 'Art']  # some missing/empty entries
})

# vertical concatenation and normalize empty strings to NaN so pandas considers them as missing
result3= pd.concat([df1, df2], axis=0, ignore_index=True)
 

result3
```
